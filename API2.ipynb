```python
import pandas as pd
import json
import requests

# 1. Cargar dataset de prueba
data = pd.read_csv("prueba.csv")

# 2. Eliminar columna de √≠ndice si existe
if "Unnamed: 0" in data.columns:
    data = data.drop("Unnamed: 0", axis=1)

# 3. Guardar los valores reales de bancarrota
y_true = data["Bankrupt?"].values.tolist()

# 4. Preparar X para la API
X_api = data.drop("Bankrupt?", axis=1)

# 5. Formato para la API
data_dict = X_api.to_dict(orient="list")
data_json = json.dumps({"data": [data_dict]})

# 6. Leer URI del servicio
with open("uri.json", "r") as f:
    scoring_uri = json.load(f)["URI"][0]

# 7. Enviar solicitud a la API
headers = {"Content-Type": "application/json"}
response = requests.post(scoring_uri, data=data_json, headers=headers)

# 8. Mostrar resultados
if response.status_code == 200:
    y_pred = json.loads(response.json())


    result_df = X_api.copy()
    result_df["Real"] = y_true
    result_df["Predicci√≥n"] = y_pred
    display(result_df)
else:
    print(f" Error {response.status_code}: {response.text}")
```

```python

import kagglehub 
import pandas as pd
import json
import requests


path = kagglehub.dataset_download("fedesoriano/company-bankruptcy-prediction")
data = pd.read_csv(path + "prueba.csv")

# seleccion de 10 aleatorias de clase 1 y 10 de clase 0
data_1 = data[data["Bankrupt?"] == 1].sample(n=10, random_state=42)
data_0 = data[data["Bankrupt?"] == 0].sample(n=10, random_state=42)
test_data = pd.concat([data_1, data_0]).sample(frac=1, random_state=42).reset_index(drop=True)

y_true = test_data["Bankrupt?"].values.tolist()
X_api = test_data.drop("Bankrupt?", axis=1)

# API
data_dict = X_api.to_dict(orient="list")
data_json = json.dumps({"data": [data_dict]})

#  URI
with open("uri.json", "r") as f:
    scoring_uri = json.load(f)["URI"][0]


headers = {"Content-Type": "application/json"}
response = requests.post(scoring_uri, data=data_json, headers=headers)


if response.status_code == 200:
    y_pred = json.loads(response.json())
    print("Predicciones:", y_pred)
    print(" Valores reales:", y_true)

    result_df = X_api.copy()
    result_df["Real"] = y_true
    result_df["Predicci√≥n"] = y_pred
    display(result_df)
else:
    print(f"Error {response.status_code}: {response.text}")
```
